```python
def process_report(text):

  # regex masks
  x_mask_pattern = re.compile(r"X{2,}")
  punct_pattern =  re.compile(r"[^a-zA-Z0-9,\s\.]")
  num_bullets_pattern = re.compile(r"^\s*\d+\s*[\.\)]\s*",re.MULTILINE)
  tab_spacing_pattern = re.compile(r"\s+")


  # split the text into sentences
  sentences = re.split(r'(?<=[\.\?!])\s+', text)

  processed_sents = []

  for sent in sentences:
    # regex cleanup
    sent = x_mask_pattern.sub("",sent)
    sent = punct_pattern.sub(" ",sent)
    sent = num_bullets_pattern.sub("",sent)
    sent = tab_spacing_pattern.sub(" ",sent)

    if len(sent) > 2:
      processed_sents.append(sent)


  processed_text = " ".join(processed_sents).lower().strip()



  return processed_text
```

```python
def encode_clip_text(texts, model, processor, device, batch_size=16):
    all_embeddings = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]

        inputs = processor(
            text=batch,
            padding=True,
            truncation=True,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():
            text_features = model.get_text_features(**inputs)
            text_features = F.normalize(text_features, p=2, dim=1)

        all_embeddings.append(text_features.cpu().numpy())

    return np.vstack(all_embeddings)
```

### MedSpacy Entity Extraction Setup

```python
# configuring NLP pipeline
cxray_nlp = spacy.load("en_core_sci_md")
print(cxray_nlp.pipe_names)
cxray_nlp.add_pipe(
    "negex",
    config={
        "ent_types": ["ENTITY"]
    },
    last=True
)
```

```python
doc = cxray_nlp("No evidence of pneumonia or pleural effusion.")

for ent in doc.ents:
    print(ent.text, ent.label_, ent._.negex)
```

```python
# minimal cleaning that is required for NER
def minimal_clean(text):

  text = re.sub(r"\s+", " ", text)  # remove multiple spaces
  text = re.sub(r"X{2,}", "", text) # remove PHI Masking
  text = text.lower()
  text = text.strip()

  return text

```

```python
# Normalizing Labels generated through Spacy

def normalize_entity(text):
  ''' Need to normalize returned entity text to
  make it consistent for use as a label '''

  text = text.lower()
  text = text.strip()

  text = re.sub(r"[^a-z0-9\s\-]", "", text)  # keep only letters, digits, hyphen, space
  text = re.sub(r"\s+", "_", text)  # replace space with underscore

  return text

```

```python
# Extract Weak Supervision Labels through sciSpacy
# Disease whitelist Ontology-guided weak supervision

'''
Conceptual Working
normalize_entity() → “pneumonia” and “Pneumonia.” → pneumonia

ent._.negex → True = absent, False = present
'''
CXR_DISEASE_VOCAB = {
    "pneumonia", "pneumothorax", "pleural_effusion", "atelectasis",
    "cardiomegaly", "edema", "emphysema", "fibrosis", "consolidation",
    "opacity", "nodule", "mass", "infiltrate", "effusion",
    "airspace_disease", "interstitial_disease", "injury"
}

def extract_labels(text, nlp):

  clean_text = minimal_clean(text)

  doc = nlp(clean_text) # extract affirmation/negation for suspected condition

  affirmed_conditions = set()
  negated_conditions = set()

  for entity in doc.ents:
    label = normalize_entity(entity.text)

    if label not in CXR_DISEASE_VOCAB:
      continue

    if len(label) < 4:
      continue

    if entity._.negex:
      negated_conditions.add(label)
    else:
      affirmed_conditions.add(label)

  return affirmed_conditions, negated_conditions

```

```python
# Apply function to Report

affirmed_condition = []
negated_condition = []

for text in raw_report['diagnosis']:
  affirmed, negated = extract_labels(text, cxray_nlp)
  affirmed_condition.append(affirmed)
  negated_condition.append(negated)

raw_report['affirmed_condtions'] = affirmed_condition
raw_report['negated_condition'] = negated_condition

raw_report.head()
```

```python
import torch
import os, glob
from PIL import Image
import skimage.io
import torchvision.transforms as T
from torchvision.transforms import InterpolationMode
import numpy as np
from torch.utils.data import Dataset, DataLoader
```

```python

class XrayDataset(Dataset):

  def __init__(self, ROOT_PATH):

    self.ROOT_PATH = ROOT_PATH
    self.image_paths = sorted(glob.glob(os.path.join(ROOT_PATH, '*.png')))

    self.pre_process = T.Compose([
    T.Resize((224,224), interpolation=InterpolationMode.BILINEAR),
    T.ToTensor(),
    T.Lambda(lambda x: x * 2.0 - 1.0),
    ]
    )

    self.center_crop = xrv.datasets.XRayCenterCrop()
    self.resizer = xrv.datasets.XRayResizer(224)

  def _extract_uid(self, path):
    img_name = path.split('/')[-1]
    uid = int(img_name.split('_')[0])
    return uid

  def __len__(self):
    return(len(self.image_paths))

  def __getitem__(self, index):

    img_path = self.image_paths[index]
    uid = self._extract_uid(img_path)

    image = Image.open(img_path).convert('L')
    image_tnsr = self.pre_process(image)

    return uid, img_path, image_tnsr
```

```python
def generate_image_embeds(image_encoder, batch_images):

  # Features
  features = image_encoder.features(batch_images)

  # Average Pooling to Remove Spatial Grid Info
  pooled_features = F.adaptive_avg_pool2d(features,(1,1) )

  # Flatten Pooled Feature
  flat_features = pooled_features.view(pooled_features.size(0), -1)

  # Normalization
  normalized_features = F.normalize(flat_features, dim=1)

  return normalized_features
```

### Building Re Ranker

```python
import numpy as np

def l2_normalization(X, eps=1e-12):
    X = np.asarray(X, dtype=np.float32)
    if X.ndim == 1:
        return X / (np.linalg.norm(X) + eps)
    return X / (np.linalg.norm(X, axis=1, keepdims=True) + eps)
```

```python
# Consensus clustering
from sklearn.cluster import AgglomerativeClustering

def rerank_retrieved_embeds(retrieved_cases,
                       text_embed_col = 'diagnosis_embeds',
                       top_k = 5,
                       min_cluster_size = 3,
                       dist_threshold = 0.03):

  embeds = np.vstack(retrieved_cases[text_embed_col].values).astype('float32')
  embeds = l2_normalization(embeds)

  cos_sim = embeds @ embeds.T
  distance = 1.0 - cos_sim
  np.fill_diagonal(distance, 0.0)

  clustering = AgglomerativeClustering(
        metric="precomputed",
        linkage="average",
        distance_threshold=dist_threshold,
        n_clusters=None
    )

  labels = clustering.fit_predict(distance)

  result = retrieved_cases.copy()
  result['cluster'] = labels

  result['intra_clust_sim'] = cos_sim.mean(axis=1)

  cluster_stats = (
        result.groupby("cluster")
           .agg(
               size=("cluster", "size"),
               avg_intra=("intra_clust_sim", "mean")
           )
           .sort_values(["size", "avg_intra"], ascending=False)
    )

  #Fallback if clustering fragments too much
  if cluster_stats.iloc[0]["size"] < min_cluster_size:
        print('Clustering fragment too much, falling back to img score')
        return (
            result.sort_values("sim_score", ascending=False)
               .head(top_k)
               .reset_index(drop=True),
            cluster_stats
        )

  best_cluster = cluster_stats.index[0]

  best = (
        result[result["cluster"] == best_cluster]
        .sort_values("intra_clust_sim", ascending=False)
        .head(top_k)
        .reset_index(drop=True)
    )

  return best, cluster_stats
```

```python
# minimal cleaning that is required for NER
def minimal_clean(text):

  text = re.sub(r"\s+", " ", text)  # remove multiple spaces
  text = re.sub(r"X{2,}", "", text) # remove PHI Masking
  text = text.lower()
  text = text.strip()

  return text
```

```python
def normalize_entity(text):
  ''' Need to normalize returned entity text to
  make it consistent for use as a label '''

  text = text.lower()
  text = text.strip()

  text = re.sub(r"[^a-z0-9\s\-]", "", text)  # keep only letters, digits, hyphen, space
  text = re.sub(r"\s+", "_", text)  # replace space with underscore

  return text
```

```python
def extract_labels(text, nlp):

  clean_text = minimal_clean(text)

  doc = nlp(clean_text) # extract affirmation/negation for suspected condition

  affirmed_conditions = set()
  negated_conditions = set()

  for entity in doc.ents:
    label = normalize_entity(entity.text)

    if label not in CXR_DISEASE_VOCAB:
      continue

    if len(label) < 4:
      continue

    if entity._.negex:
      negated_conditions.add(label)
    else:
      affirmed_conditions.add(label)

  return affirmed_conditions, negated_conditions
```

```python
# Affirmed Recall (Coverage): Did model cover true finding
def affirmed_recall(PS_aff, GT_aff):
  if len(GT_aff) == 0:
    return 1.0
  return len(PS_aff & GT_aff)/len(GT_aff)

# Affirmed Precision (Correctness):
def affirmed_precision(PS_aff, GT_aff):
  if len(PS_aff) == 0:
    return 1.0
  return len(PS_aff & GT_aff) / len(PS_aff)

# Polarity Flip Rate: Did we contrdict GT
def polarity_flip_count(PS_aff, GT_ng):
  return len(PS_aff & GT_ng)

# Evidence Support Rate: Was the output grounded
def evidence_support(PS_aff, ES_aff, ES_ng):
  if len(PS_aff) == 0:
    return 1.0
  supported = ES_aff | ES_ng
  return len(PS_aff & supported) / len(PS_aff)

# Hallucination wrt Ground Truth
def hallucination_wrt_gt(PS_aff, GT_aff):
  if len(PS_aff) == 0:
    return 0.0
  return len(PS_aff - GT_aff) / len(PS_aff)

#Hallucination wrt Evidence Set
def hallucination_ert_evd(PS_aff, ES_aff, ES_ng):
  if len(PS_aff) == 0:
    return 0.0
  supported = ES_aff | ES_ng
  return len(PS_aff - supported) / len(PS_aff)

# Negated Recall: Did model find normal findings
def negated_recall(PS_ng, GT_ng):
    if len(GT_ng) == 0:
        return 1.0
    return len(PS_ng & GT_ng) / len(GT_ng)

# Negated Precision
def negated_precision(PS_ng, GT_ng):
    if len(PS_ng) == 0:
        return 1.0
    return len(PS_ng & GT_ng) / len(PS_ng)
```

```python
from jinja2 import Template
XRAY_REPORT_TEMPLATE = Template (
"""
AI-Assisted Clinical Decision Support System
CHEST X-RAY SUGGESTIVE DIAGNOSIS REPORT


Patient Name: {{ patient_name }}
Age: {{ age }}


IMPRESSION:
{{ impression }}


Generated by: {{ generator_name }}


This report is AI-assisted and intended for research purposes only.
""".strip()
)
```

```python
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors
from reportlab.lib.enums import TA_CENTER
from reportlab.platypus import Image, Spacer, Paragraph
from reportlab.lib.units import inch
from reportlab.platypus import SimpleDocTemplate
from PIL import Image as PILImage
from reportlab.platypus import HRFlowable
from reportlab.lib import colors
from reportlab.lib.pagesizes import LETTER

def render_report(data):
  return XRAY_REPORT_TEMPLATE.render(
      patient_name = data['patient_name'],
      age = data['age'],
      impression = data['impression'],
      generator_name = 'Visaul RAG System'
 )


def save_report_pdf_v2(report_text: str, cxr_image_path, out_path: str):

    styles = getSampleStyleSheet()

    header_style = ParagraphStyle(
        'HeaderStyle',
        parent=styles['Normal'],
        fontSize=16,
        leading=20,
        alignment=TA_CENTER,
        spaceAfter=12,
    )

    body_style = ParagraphStyle(
        'BodyStyle',
        parent=styles['Normal'],
        fontSize=12,
        leading=16,
    )

    doc = SimpleDocTemplate(
        out_path,
        pagesize=LETTER,
        rightMargin=54,
        leftMargin=54,
        topMargin=54,
        bottomMargin=54
    )

    story = []

    blocks = report_text.split("\n\n")

    for i, block in enumerate(blocks):
        block = block.strip().replace("\n", "<br/>")

        if block:
            if i == 0:
                story.append(Paragraph(f"<b>{block}</b>", header_style))
                story.append(HRFlowable(
                          width="100%",
                          thickness=1,
                          lineCap='round',
                          color=colors.black,
                          spaceBefore=6,
                          spaceAfter=6
                      ))
            elif i == len(blocks) - 1:
              story.append(HRFlowable(
                                    width="100%",
                                    thickness=1,
                                    lineCap='round',
                                    color=colors.black,
                                    spaceBefore=6,
                                    spaceAfter=6
                                ))
              story.append(Paragraph(block, body_style))
            else:
                story.append(Paragraph(block, body_style))

            story.append(Spacer(1, 20))


        if i == 0 and cxr_image_path:
            pil_img = PILImage.open(cxr_image_path)
            w, h = pil_img.size
            aspect = h / w

            max_w = 6.0 * inch
            img = Image(cxr_image_path, width=max_w, height=max_w * aspect)

            story.append(Spacer(1, 8))
            story.append(img)
            story.append(Spacer(1, 14))



    doc.build(story)
```
